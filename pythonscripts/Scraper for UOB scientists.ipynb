This is going to be a scraper that will scrape the names and descriptions of scientists at UoB in the Science of the Virus. 

So first we are going to install the packages we need etc

```python
!pip install scraperwiki
import scraperwiki

import lxml.html

!pip install cssselect
import cssselect

import pandas as pd
```

So we have all of our packages that we need installed, lets begin with trying to get that data! 

```python
uoburl = "https://www.birmingham.ac.uk/research/coronavirus/find-an-expert.aspx"

#We are now create a dataframe to store the data we are about to scrape.
df = pd.DataFrame(columns=["title","description"])

#Now lets scrape that badboi
html = scraperwiki.scrape(uoburl)

#Lets turn this into an lxml instead of a HTML
root = lxml.html.fromstring(html)

#Locate where the information is stored in the webpage you want to scrape
titles = root.cssselect('article ul li a')
descs = root.cssselect('article ul li')

#we are going to find out the list length. For some reason, descriptions works and titles doesnt, but we won't question that right now!
listlength = len(descs)
print(listlength)

#lets loop that through and see what we get and make sure it matches with the number of entries on the website!
for i in range(0,listlength):
    print(i)
    
#this makes all the gobbledegloop data readable!    
    title = titles[i].text_content()
    desc = descs[i].text_content()
    print(title, desc)
    
#lets put all of our data into a dataframe to make it better    
    df = df.append({
    "title" : title,
    "desc" : desc
    }, ignore_index=True)

    print(df)
```

Woohoo! We got it working! We're now going to print that dataframe so that we can see the data.

```python
print(df)
```
